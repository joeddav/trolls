```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import glob
sns.set()
%config InlineBackend.figure_format = 'retina'
```

#### Get data
Data stored in 9 csv files in `~/data`

```python
DATA = 'data'
allFiles = glob.glob(DATA + "/*.csv")
list_ = []
for file_ in allFiles:
    df = pd.read_csv(file_)
    list_.append(df)
data = pd.concat(list_)
```

```python
data.head()
```

#### Category Types
Troll tweets run into one of several categories. Here's the breakdown:

```python
cat_counts = data.account_category.value_counts()
```

```python
_, cat_ax = plt.subplots(figsize=(8,8))
cat_ax.pie(cat_counts, labels=cat_counts.index)
plt.title('Breakdown of tweet "Categories"')
plt.show()
```

```python
rt_counts = data.retweet.value_counts()
```

```python
_, rt_ax = plt.subplots(figsize=(8,8))
rt_ax.pie(rt_counts, labels=['Organic', 'Retweet'])
plt.title('Retween Proportion')
plt.show()
```

```python
rgn_counts = data.region.value_counts()[2:]
```

```python
_, rgn_ax = plt.subplots(figsize=(16,8))
rgn_ax.bar(rgn_counts.index, rgn_counts.values)
plt.xticks(rotation=90)
plt.title('Top User Regions after "USA" and "Unknown")')
plt.show()
```
